```python
class Luhn:
    def __init__(self, card_num):
        self.luhn = card_num

    def valid(self):
        if len(self.luhn) <= 1:
            return ("Not Valid")
        else:
            # Remove spaces.
            clean_card_num = self.luhn.replace(" ","")
            # In case of other special characters return "Not Valid".
            try:
                card_num = int(clean_card_num)
            except:
                return ("Not Valid")
            # Reverse card number
            rev_card_num = clean_card_num[::-1]
            upd_card_num = ""
            check_sum = 0
            # Double every second digit and sum all digits.
            for i in range(len(rev_card_num)):
                digit = int(clean_card_num[i])
                if (i % 2) != 0:
                    digit = (digit * 2) % 9
                upd_card_num += str(digit)
                check_sum += digit
                print ("upd_card_num + Accumulated sum: ", upd_card_num, " ", check_sum)
            # If the sum is evenly divisible by 10, then the number is valid
            if check_sum % 10 != 0:
                return ("Not Valid")
            else:
                return ("Valid")

# Program tester
while True:
    card_num = input ("Card Number: ")
    if card_num == "done":
        break
    luhn_card_num = Luhn(card_num)
    print ("Luhn: ", luhn_card_num.luhn)
    print ("Valid? ", luhn_card_num.valid())

print ("=> End of program")
```
